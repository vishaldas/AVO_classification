```python
# AVO Parameters
src_len=61
freq=40
dt=0.002
thetaStart=0
thetaEnd=30
thetaInc=1
theta=np.arange(thetaStart,thetaEnd+thetaInc,thetaInc)
ntheta=theta.size
nt=61
ncdp=1111
angle=np.reshape(theta,[1,ntheta])

time=np.arange(0,nt,1)*dt
```

```python
# Generate the source wavelet
def ricker(f,n,dt,t0):

    T = dt*(np.floor(n/2));
    t = np.linspace(-T,T,n)
    tau = t-t0;
    pi=np.pi
    s = (1-tau*tau*f**2*pi**2)*np.exp(-tau**2*pi**2*f**2);
    rw = s;

    return rw,t
src,tsrc=ricker(freq,src_len,dt,0)
plt.plot(tsrc,src)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Source Wavelet")
plt.show()
```

```python
# Generate gathers for Class 1 AVO

R_class1 = np.zeros((R0_G_sampled_class1.shape[0], len(theta)))
for i in range(len(theta)):
    R_class1[:,i] = R0_G_sampled_class1[:,0] + R0_G_sampled_class1[:,1]*np.square(np.sin(np.deg2rad(theta[i])))
    
# Generating angle gathers
angle_gather_class1 = np.zeros((R0_G_sampled_class1.shape[0], len(src), len(theta)))
for ii in range(R0_G_sampled_class1.shape[0]):
    for jj in range(len(theta)):
        angle_gather_class1[ii,:,jj] = np.convolve(R_class1[ii,jj], src, 'same')
```
